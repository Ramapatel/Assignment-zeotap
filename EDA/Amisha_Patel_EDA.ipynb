```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
products=pd.read_csv('Products.csv')
transactions=pd.read_csv('Transactions.csv')
customers=pd.read_csv('Customers.csv')
```

```python
#Checking missing values are not present or not
print("Products Dataset:\n")
print(products.isnull().sum())
print("Transaction Dataset:\n")
print(transactions.isnull().sum())
print("Customer Dataset: \n")
print(customers.isnull().sum())
```

```python
# Data Information
print(products.info())
print(transactions.info())
print(customers.info())
```

```python
# Popular Products
top_products = transactions['ProductID'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sn.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 10 Most Purchased Products")
plt.show()
```

```python
# Customers by Region
plt.figure(figsize=(8, 5))
sn.countplot(data=customers, x='Region')
plt.title("Customers by Region")
plt.show()
```

```python
# Heatmap for Transaction
numeric_cols = transactions.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sn.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```
